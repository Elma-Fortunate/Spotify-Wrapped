```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import pandas as pd
import time

CLIENT_ID = "528de54c3cd046bc90499ce4c2c45dc3"
CLIENT_SECRET = "298cc40b62184ead92189451bc1b9220"

sp = spotipy.Spotify(
    auth_manager=SpotifyClientCredentials(
        client_id=CLIENT_ID,
        client_secret=CLIENT_SECRET
    )
)
```

```python
spotify_25["track_id"] = spotify_25["spotify_track_uri"].astype(str).str.extract(r"spotify:track:([A-Za-z0-9]+)")
spotify_25["track_id"]
```

```python
spotify_25["track_id"].nunique()
```

```python
import time

cover_map = {}
track_ids = spotify_25["track_id"].dropna().unique().tolist()
TOTAL = len(track_ids)
print("Total tracks to process:", TOTAL)

for i in range(0, TOTAL, 50):  # Spotify allows 50 tracks per request
    batch = track_ids[i:i+50]

    try:
        response = sp.tracks(batch)
        tracks = response["tracks"]

        for track in tracks:
            if track and track["album"] and track["album"]["images"]:
                cover_map[track["id"]] = track["album"]["images"][0]["url"]  # highest resolution
            else:
                cover_map[track["id"]] = None

        print(f"Processed {min(i+50, TOTAL)} / {TOTAL}")
        time.sleep(1)  # safety pause to avoid rate limits

    except Exception as e:
        print("Error at batch:", i, e)
        time.sleep(10)
```

```python
spotify_25["cover_url"] = spotify_25["track_id"].map(cover_map)

```
